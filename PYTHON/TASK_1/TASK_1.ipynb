```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df_wtht_nan = df.dropna(subset=["Embarked"])
df_wtht_nan.info()
```

```python
df_wtht_nan["Age"].fillna(df["Age"].mean(), inplace=True)
```

```python
df_wtht_nan.info()
```

```python
mode_сabin = df['Cabin'].mode()
mode_сabin
```

```python
mode_сabin[0]
```

```python
cabin_fill = mode_сabin[0].split()[0]
cabin_fill
```

```python
df_wtht_nan["Cabin"].fillna(cabin_fill, inplace=True)
```

```python
df_wtht_nan.info()
```

```python
df_wtht_nan.describe()
```

```python
df["Sex"].unique()
```

```python
df_male = df_wtht_nan[df_wtht_nan["Sex"] == "male"]
df_female = df_wtht_nan[df_wtht_nan["Sex"] == "female"]
```

```python
import matplotlib.pyplot as plt
```

Для части датасета male

```python
df_male.describe()
```

```python
age_hist = df_male["Age"].hist(bins=10, edgecolor="black", grid=False)
plt.title("Распределение возраста")
plt.xlabel("Возраст")
plt.ylabel("Частота")
```

```python
plt.boxplot(df_male["Age"], vert=False, patch_artist=True,
            boxprops=dict(facecolor="lightblue"))
plt.title("Boxplot возраста")
plt.xlabel("Возраст")
```

```python
plt.violinplot(df_male["Age"], vert=False, showmedians=True)
plt.title("Violinplot возраста")
plt.xlabel("Возраст")
```

```python
plt.scatter(range(len(df_male["Age"])), df_male["Age"], color="blue", 
            alpha=0.5)
plt.title("Scatter plot возраста")
plt.xlabel("Частота")
plt.ylabel("Возраст")
```

```python
age_bins = [20, 30, 40, 50, 60, 70, 80]
age_groups = pd.cut(df_male["Age"], bins=age_bins)
age_group_counts = age_groups.value_counts().sort_index()
age_group_counts.plot(kind="bar", color="orange", edgecolor="black")
plt.title("Bar plot возраста")
plt.xlabel("Возрастные группы")
plt.ylabel("Количество")
```

```python
age_group_counts.plot(kind="pie", autopct='%1.1f%%', startangle=90, colors=
                      ["lightblue", "lightgreen", "lightcoral", "lightskyblue", 
                       "lightpink"])
plt.title("Pie chart возрастных групп")
```
